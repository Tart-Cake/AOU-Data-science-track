```python
# So to deal with the missing values in Age and Cabin there are serval ways to do this however
# changing values isn't an easy task because changing values is changing in the information you have
# so I will try to be carful in that aspect for the age I will consider leave it and use it.
# for the cabin I will not change any value I will just make another column that has the value hasCabin.
```

```python
df["HasCabin"] = df["Cabin"].notna().astype(int)
```

```python
df
```

```python
df["HasCabin"].replace(0,'NO',inplace=True)
df["HasCabin"].replace(1,'YES',inplace=True)
df
```

```python
df.isnull().sum()
```

```python
# First I wnat to know the reason behind the missing information, I will try to see if the PClass has somehing with it.
print("Missing age values by Pclass:")
for pclass in df['Pclass'].unique():
    missing_age_count = df[df['Pclass'] == pclass]['Age'].isnull().sum()
    total_passengers = df[df['Pclass'] == pclass].shape[0]
    percentage_missing = (missing_age_count / total_passengers) * 100
    print(f"  - Pclass {pclass}: {missing_age_count} missing values ({percentage_missing:.2f}%)")
```

```python
# Chart to represent the percentage of the gender distribution on the ship.
gender_count = df["Sex"].value_counts()
gender_count
plt.axis("equal")
plt.pie(gender_count, labels = ["Male","Female"],autopct='%0.0f%%')
plt.legend(title="Gender")
plt.show()
```

```python
# Chart to represent the percentage of the survival distribution on the ship.
survived_counts = df["Survived"].value_counts()
survived_counts
plt.axis("equal")
plt.pie(survived_counts, labels = ["Dead","Alive"],autopct='%0.0f%%')
plt.legend(title="Survival")
plt.show()
```

```python
# Chart to represent the percentage of the passengers class distribution on the ship.
pClass_counts = df["Pclass"].value_counts()
pClass_counts
plt.pie(pClass_counts, labels=["Third Class","First Class","Second Class"],autopct='%0.0f%%')
plt.legend(title="Passengers Class")
plt.show()
```

```python
# Chart to represent the Survival probability by the passenger class. 
survived_counts = df.groupby("Pclass")["Survived"].value_counts(normalize=True)
survived_counts.unstack().plot(kind="bar")
plt.xlabel("Passenger class")
plt.ylabel("Survival Probability (1: Yes, 0: No)")
plt.title("Survival Probability by Passenger Class")
plt.legend(title="Survived")
plt.grid(True)
plt.show()
```

```python
# Chart to represent the Age Distributio on the ship.
plt.figure(figsize=(8, 6))
plt.hist(df["Age"].dropna(), bins=20, edgecolor="black")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")
plt.title("Age Distribution")
plt.grid(True)
plt.show()
```

```python
# Chart to represent the Age and survival probability 
# Calculate survival rate as proportion for all ages
survival_rate = df.groupby("Age")["Survived"].sum() / df.groupby("Age").size()

# Define age groups (adjust ranges as needed)
age_groups = [(0, 10), (11, 20), (21, 30), (31, 40), (41, 50), (51, 60), (61, 70), (71, 80)]

# Calculate survival rate for each group
group_survival_rates = [
    survival_rate[(survival_rate.index >= start) & (survival_rate.index <= end)].mean()
    for start, end in age_groups
]

# Create clear labels for groups
group_labels = [f"{start}-{end}" for start, end in age_groups]

# Create the bar chart with informative labels and annotations
plt.figure(figsize=(10, 6))  # Set a suitable figure size
plt.bar(group_labels, group_survival_rates, color='skyblue')  # Create blue bars
plt.xlabel("Age Group", fontsize=12)  # Set x-axis label with larger font
plt.ylabel("Survival Rate (%)", fontsize=12)  # Set y-axis label with larger font
plt.title("Proportion of Survivors by Age Group (All Ages)", fontsize=14)  # Set informative title
plt.grid(True)  # Enable grid lines

# Add data labels (percentages) above bars for clarity
for i, rate in enumerate(group_survival_rates):
    plt.text(i, rate + 0.01, f"{rate:.1%}", ha='center', va='bottom', fontsize=10)

plt.show()  # Display the chart
```

```python
# Chart to represent the missing age values and the survived passengers 
# Count passengers with missing age and those survived/not survived
age_missing = df[df['Age']==np.nan]
age_missing_count = df['Age'].isna().sum()
survived_count = df[df['Survived'] == 1].shape[0]
not_survived_count = df[df['Survived'] == 0].shape[0]

# Create pie chart with labels
slices = ['Missing Age', 'Survived', 'Not Survived']
colors = ['lightgray', 'green', 'red']
sizes = [age_missing_count, survived_count, not_survived_count]

plt.pie(sizes, labels=slices, autopct="%1.1f%%", startangle=90, colors=colors)
plt.title('Missing Age Values and Survival Rates')
plt.axis('equal')  # Equal aspect ratio ensures a circular pie chart
plt.show()
```
